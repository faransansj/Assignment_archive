```python
import numpy as np
import matplotlib.pyplot as plt
from math import *

np.random.seed(6974)
N =1

print(np.random.rand(N))
data = np.array(np.random.rand(N))

print(data)

y, binEges = np.histogram(data, bins=2)
bincenters = 0.5*(binEges[1:]+binEges[:-1])
width = 0.49

errors = np.sqrt(y)

plt.bar(bincenters, y, width=width, yerr=errors)
plt.show()
```
