## ✅ 실습과제 6-1: 행렬의 인덱싱과 슬라이싱

```python
import numpy as np

x = np.array([
    [1, 0, 1, 1],
    [0, 2, 1, -1],
    [1, 3, 4, -1]
])

print("문제 1:", x[2, 1])      # 3행 2열
print("문제 2:", x[1, :])      # 2행 전체
print("문제 3:", x[:, 2])      # 모든 행의 3열
print("문제 4:", x[1, 1:4])    # 2행 2열~4열
```

## ✅ 실습과제 6-2: 행렬 곱셈

```python
A = np.array([
    [1, 0, 1, 1],
    [0, 2, 1, -1],
    [1, 3, 4, -1]
])

B = np.array([
    [1, 0, 1],
    [0, 2, 1],
    [1, 3, 4],
    [0, 2, 1]
])

result = np.dot(A, B)
print(result)
```

## ✅ 실습과제 6-3: 정규분포 난수 생성 및 행별 합

```python
x = np.random.normal(loc=0, scale=0.2, size=(4, 10))
print("x:\n", x)
print("각 행의 합:", np.sum(x, axis=1))
```

## ✅ 실습과제 6-4: 다항 함수 시각화

```python
import matplotlib.pyplot as plt

x = np.linspace(-3, 3, 400)
f = x * (x + 1) * (x - 1) * (x - 2)
g = x * (x + 1) * (x + 2) * (x - 1)

plt.plot(x, f, label='f(x)')
plt.plot(x, g, label='g(x)')
plt.grid()
plt.legend()
plt.show()
```

## ✅ 실습과제 6-5: 선 굵기 및 범례

```python
x = np.linspace(0, 10, 100)
y1 = x
y2 = x ** 2

plt.plot(x, y1, 'b', linewidth=3, label='line1-width-3')
plt.plot(x, y2, 'r', linewidth=5, label='line2-width-5')
plt.legend()
plt.show()
```

## ✅ 실습과제 6-5-1: 3차 다항함수 시각화

```python
x = np.linspace(-10, 10, 400)
y = x**3 + 3*x**2 + 2

plt.plot(x, y, 'r--', label='order 3')
plt.title('graph of polynomial of degree 3')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.grid()
plt.show()
```
